```python
import requests
import json
import operator
import re

import pandas as pd
from rdkit import Chem as Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
IPythonConsole.ipython_useSVG=True  #< set this to False if you want PNGs instead of SVGs

def submit_query(query: str, length: int, qopts: str, search_type: str, db_choice: str):
    
    #search_type= ['SMI', 'SUB', 'SMA']
                        
    # database choice list is updated with ``df = SmallWorld.retrieve_databases()
    #db_choices = ['BB-10-22Q1','BB-20-22Q1','BB-30-22Q1','BB-40-22Q1','BB-50-22Q1','BB-Stock-19Q4-1.6M','BB-All-For-Sale-19Q4-26M','HMDBMetabolites-20Q1-585','In-Stock-19Q4-13.8M','Mcule_ultimate_20Q2-126M','REAL_Database_21Q3_2B','TCNMP-20Q1-37K','Wait-OK-19Q4-1.1B','World-Drugs-20Q1-3K','WuXi-19Q4-339M','ZINC-All-19Q4-1.4B','ZINC-Interesting-2019Q4-307K','ZINC-On-Demand-19Q4-311M','ZINC20-ForSale-22Q1']

    # most of these are enabled by default
    #qopts= ['','R','C','RC']
        
        
    url = "http://arthor.docking.org/dt/"
    url = url + db_choice
    url = url + "/search?query="
    
    url = url + query
    url = url + "&type="    
    url = url + search_type 
    url = url + "&start=0&length="
    url = url + str(length)

    print(url)
    
    resp = requests.get(url)
    jsonResponse = resp.json()
    
    hitSmiles = []
    for data in jsonResponse['data']:
        hitSmiles.append(data[1].split()[0])
    
    return(hitSmiles)
```

```python
#query = 'CC(=O)OC1=CC=CC=C1C(=O)O'
query  ='CN2CCN(C1=C(Cl)C(Cl)=CC=C1)CC2'
max_hits = 25000
fix = 'C' #R fix rings C:fix chains
search_type = 'SUB' # SUB, SIM or SMA
library = 'REAL_Database_21Q3_2B'
hitSmiles = submit_query(query, max_hits, fix, search_type, library)

print(len(hitSmiles))

#hitSmiles = submit_query('CC(=O)OC1=CC=CC=C1C(=O)O', 250, '', 'SUB', 'Wait-OK-19Q4-1.1B')
ms = [Chem.MolFromSmiles(d) for d in hitSmiles]
Draw.MolsToGridImage(ms,molsPerRow=10,subImgSize=(200, 200),maxMols=250)
    
```
